## Step 1:

* matplotlib.pyplot 
* numpy


Note: plt and np are aliases of respective packages

```python
#use this while using jupyter notebooks or Google Colab
%matplotlib inline  

import matplotlib.pyplot as plt

import numpy as np
```

## Step 2

### Create synthetic data for the illustration
* Generate some sequence data, followed by sine and cosine values of the data.
* In this case, create 100 values between zero to 10 starting at zero.

Note: You may read data from your data sources such as sql database, ms excel, csv & other files 

```python

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

## Step 4
* Invoke the subplots object by passing data using plot(x,y) function 
  * where `x` is data on `x-axis` and `y` is data on `y-axis` of plot respectively

```python
# Plot some data on the axes.
fig, ax = plt.subplots()
ax.plot(x,y)  
```

## Step 5
* Set the title of plot using `set_title("Titile of the Plot")`

```python
# Set a title for the plot
ax.set_title('Plot of Sin(x)')
```

```python
# Plot some data on the axes.
fig, ax = plt.subplots()
ax.plot(x,y)  

# Set a title for the plot
ax.set_title('Plot of x and sin(x)')
```

```python
# Plot some data on the axes.
fig, ax = plt.subplots()
ax.plot(x,y)  

# Set a title for the plot
ax.set_title('Plot of x and sin(x)')

# Set labels for x-axis and y-axis
ax.set_xlabel('X-axis: Value of X')
ax.set_ylabel('Y-axis: Value of Sin(X)')
```

# Combine all the code into one cell 

```python
# Plot some data on the axes.
fig, ax = plt.subplots()
ax.plot(x,y)  

# Set a title for the plot
ax.set_title('Plot of x and sin(x)')

# Set labels for x-axis and y-axis
ax.set_xlabel('X-axis: Value of X')
ax.set_ylabel('Y-axis: Value of Sin(X)')

# Add a grid
ax.grid(True)
```

# Similar illustration with two sub-plots

```python
# Create just a figure and  two subplots
fig, (ax1, ax2) = plt.subplots(1,2)
```

```python
# Create just a figure and only two subplots
fig, (ax1, ax2) = plt.subplots(1,2)

# unpack the subplots
ax1.plot(x, y)
ax1.set_title('x and sin(x)')
# Set labels for x-axis and y-axis
ax1.set_xlabel('X-axis: Value of X')
ax1.set_ylabel('Y-axis: Value of Sin(X)')

ax2.plot(x, z)
ax2.set_title('x and cos(x)')
# Set labels for x-axis and y-axis
ax2.set_xlabel('X-axis: Value of X')
ax2.set_ylabel('Y-axis: Value of Cos(X)')
```

# Similar illustration with two variables on y-axis and legend

```python
# Add a legend
# Legend explians the elements displayed in a plot. 
ax.legend()
```

```python
# Create just a figure and only two subplots
#fig, (ax1, ax2, ax3) = plt.subplots(1,3)

fig, ax3 = plt.subplots()

ax3.plot(x, y, label ="Sine of numbers")
ax3.plot(x, z, label ="Cosine of numbers")

# Add a legend
# Legend explians the elements displayed in a plot. 
ax3.legend()


ax3.set_title('x, sin(x) and cos(x)')
# Set labels for x-axis and y-axis
ax3.set_xlabel('X-axis: Value of X')
ax3.set_ylabel('Y-axis: Value of Sin(X) and Cos(X)')

```
